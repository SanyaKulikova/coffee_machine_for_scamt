```python
class CoffeeMachine:   
    
    def get_input(self):  # в задании просили сделать метод для принимания инпута
        # но вторую часть задания я не очень поняла(буду рада потом обсудить)
        return input()
    
    def __init__(self, water_amount, milk_amount, coffee_amount, cups_amount, money):
        self.water_amount = water_amount
        self.milk_amount = milk_amount
        self.coffee_amount = coffee_amount
        self.cups_amount = cups_amount
        self.money = money
        while True:
            print("Write action (buy, fill, take, remaining, exit):")
            action = self.get_input()
            if action == "buy":
                self.buy()
            elif action == "fill":
                self.fill()
            elif action == "take":
                self.take()
            elif action == "remaining":
                print()
                self.show_statement()
            elif action == "exit":
                break
    def show_statement(self):
        print("The coffee machine has:")
        print("{} ml of water".format(self.water_amount))
        print("{} ml of milk".format(self.milk_amount))
        print("{} g of coffee beans".format(self.coffee_amount))
        print("{} disposable cups".format(self.cups_amount))
        money_to_string = "$" + str(self.money)
        print("{} of money".format(money_to_string))
        print()
    
    def check_bool(self, coffee_type):
        result = False
        if coffee_type == "1":
            if self.water_amount >= 250 and self.coffee_amount >= 16 and self.cups_amount >= 1:
                result = True
            return result
        if coffee_type == "2":
            if self.water_amount >= 350 and  self.milk_amount >= 75 and self.coffee_amount >= 20 and self.cups_amount >= 1:
                result = True
            return result
        if coffee_type == "3":
            if self.water_amount >= 200 and self.milk_amount >= 100 and self.coffee_amount >= 12 and self.cups_amount >= 1:
                result = True
            return result
    
    def check_less(self, coffee_type):
        less = ""
        if coffee_type == "1":
            if self.water_amount < 250:
                less = "water"
            elif self.coffee_amount < 16:
                less = "coffee beans"
            elif self.cups_amount < 1:
                less = "disposable cups"
            return less
        if coffee_type == "2":
            if self.water_amount < 350:
                less = "water"
            elif self.milk_amount < 75:
                less = "milk"
            elif self.coffee_amount < 20:
                less = "coffee beans"
            elif self.cups_amount < 1:
                less = "disposable cups"
            return less
        if coffee_type == "3":
            if self.water_amount < 200:
                less = "water"
            elif self.milk_amount < 100:
                less = "milk"
            elif self.coffee_amount < 12:
                less = "coffee beans"
            elif self.cups_amount < 1:
                less = "disposable cups"
                return less
        
    def buy(self):

        print()
        print("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino, back - to main menu: ")
        coffee_type = self.get_input()

        if coffee_type == "1":
            if self.check_bool("1"):
                print("I have enough resources, making you a coffee!")
                print()
                self.water_amount -= 250
                self.coffee_amount -= 16
                self.money += 4
                self.cups_amount -= 1
                return self.water_amount, self.milk_amount, self.coffee_amount, self.cups_amount, self.money
            else:
                print("Sorry, not enough {}".format(self.check_less("1")))
                print()
                     
        elif coffee_type == "2":
            if self.check_bool("2"):
                print("I have enough resources, making you a coffee!")
                print()
                self.water_amount -= 350 
                self.milk_amount -= 75 
                self.coffee_amount -= 20
                self.cups_amount -= 1
                self.money += 7
                return self.water_amount, self.milk_amount, self.coffee_amount, self.cups_amount, self.money
            else:
                print("Sorry, not enough {}".format(self.check_less("2")))
                print()
    
        elif coffee_type == "3":
            if self.check_bool("3"):
                print("I have enough resources, making you a coffee!")
                print()
                self.water_amount -= 200 
                self.milk_amount -= 100 
                self.coffee_amount -= 12
                self.cups_amount -= 1
                self.money += 6
                return self.water_amount, self.milk_amount, self.coffee_amount, self.cups_amount, self.money
            else:
                print("Sorry, not enough {}".format(self.check_less("3")))
                print()
        elif coffee_type == "back":
            print()
    
    def fill(self):

        print("Write how many ml of water you want to add:")
        self.water_amount += int(self.get_input())
        print("Write how many ml of milk you want to add:")
        self.milk_amount += int(self.get_input())
        print("Write how many grams of coffee beans you want to add:")
        self.coffee_amount += int(self.get_input())
        print("Write how many disposable cups you want to add:")
        self.cups_amount += int(self.get_input())
        print()
        return self.water_amount, self.milk_amount, self.coffee_amount, self.cups_amount
        
    def take(self):
        money_to_string = "$" + str(self.money)
        print("I gave you {}".format(money_to_string))
        print()
        self.money = 0
        return self.money
    

                
```
